```python
from imageai.Detection import ObjectDetection
import os
model_path="/content"
images_path="/content/img"
detector=ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(os.path.join(model_path,'yolo.h5'))
detector.loadModel()
detection=detector.detectCustomObjectsFromImage(input_image=os.path.join(images_path,'Busy-street_1000x667.jpg'),output_image_path =os.path.join(images_path,'Busy-street_p_1000x667.jpg'))


```
